```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
import pickle as pkl

# Prepare data (you would have 'ds' already defined, I assume)
X = ds[["name", "company", "year", "kms_driven", "fuel_type"]]
y = ds[["Price"]]

# Define the column transformer with OneHotEncoder for categorical columns
ct = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), ["name", "company", "fuel_type"]),  # OneHotEncoder for categorical features
    remainder="passthrough"  # Keep other columns unchanged
)

# Initialize the linear regression model
regLR = LinearRegression()

# Create a pipeline with the transformer and model
pipeLR = make_pipeline(ct, regLR)

# Initialize best score tracking
best_score = -float('inf')
best_index = -1
scores = []

# Train-test split and model evaluation
for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=i)
    pipeLR.fit(X_train, y_train)  # Fit the pipeline with training data
    y_pred = pipeLR.predict(X_test)  # Predict on test data
    score = r2_score(y_test, y_pred)  # Calculate r2 score
    scores.append(score)
    
    # Track best score and index
    if score > best_score:
        best_score = score
        best_index = i

# Save the trained model as a pickle file
pkl.dump(pipeLR, open("MLR.pkl", "wb"))

# Test prediction for a new input
data = ["Santro Xing XO eRLX Euro III", "Hyundai", 2010, 50000, "Petrol"]
columns = ["name", "company", "year", "kms_driven", "fuel_type"]
df = pd.DataFrame([data], columns=columns)

# Get the predicted value and round it
prediction = pipeLR.predict(df)[0, 0]  # Get scalar value from the prediction
result = round(prediction)  # Round the predicted value
print(f"Predicted result: {result}")
```
